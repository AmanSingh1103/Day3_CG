```python
import pandas as pd
from google.colab import files

uploaded = files.upload()

file_names = list(uploaded.keys())

customers = pd.read_csv(file_names[0])
sales = pd.read_csv(file_names[1])
support = pd.read_csv(file_names[2])

customers.columns = customers.columns.str.strip()
sales.columns = sales.columns.str.strip()
support.columns = support.columns.str.strip()

for col in customers.columns:
    if "date" in col.lower():
        customers[col] = pd.to_datetime(customers[col], errors="ignore")

for col in sales.columns:
    if "date" in col.lower():
        sales[col] = pd.to_datetime(sales[col], errors="ignore")

for col in support.columns:
    if "date" in col.lower():
        support[col] = pd.to_datetime(support[col], errors="ignore")

data = customers.merge(sales, on="CustomerID", how="left")
data = data.merge(support, on="CustomerID", how="left")

print("Customers Shape:", customers.shape)
print("Sales Shape:", sales.shape)
print("Support Shape:", support.shape)

print("\nFinal Merged Data:")
print(data.head())

print("\nTotal Revenue:")
if "Amount" in data.columns:
    print(data["Amount"].sum())

print("\nAverage Revenue per Customer:")
if "Amount" in data.columns:
    print(data.groupby("CustomerID")["Amount"].sum().mean())
```
