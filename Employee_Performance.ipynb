```python
import pandas as pd
from google.colab import files

uploaded = files.upload()

file_name = list(uploaded.keys())[0]

df = pd.read_csv(file_name)

df.columns = df.columns.str.strip()

print("Original Data")
print(df)

print("\nExceptional Performers (Score > 90)")
print(df[df["Score"] > 90])

print("\nSales Department Records")
print(df[df["Department"] == "Sales"])

print("\nRanked by Score (Descending)")
print(df.sort_values(by="Score", ascending=False))

print("\nDepartment-wise Ranking")
print(df.sort_values(by=["Department", "Score"], ascending=[True, False]))

print("\nAverage Score per Department")
print(df.groupby("Department")["Score"].mean().reset_index())

print("\nMaximum Score per Quarter")
print(df.groupby("Quarter")["Score"].max().reset_index())
```
